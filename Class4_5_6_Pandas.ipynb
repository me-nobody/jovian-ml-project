74.35% of females survived and 18.78% of males survived

```python
#there is a drastic difference in the survival rate of females and males
#hence a higher prob of survival for females than of males
#survived is the response variable, we understand through this, that gender is leading to huge variance in the survival rate,
#and hence in the prediction problem of survival, gender will be a key variable which will help us in prediction whether 
#the person survives or not.

#feature selection: it's important to the select the important features which will lead to a high variance in the response variable.
#this is important as we don't want unimportant features in our model, as they lead to over-fitting to the training data. 
```

```python
# sex is leading to a high variance in fare column - which means that if we have to do the missing values imputation in fare column,
#then we can also check the sex of the person and impute the missing values in the fare column by the mean fare paid by male / female respectively
```

```python
titanic.Fare=np.where(titanic.Fare.notnull(), titanic.Fare, titanic.groupby("Sex").Fare.transform("mean"))
#we could use the transform fn here, because, sex is responsible for a high variance in the average fare
```

```python
data = {'raw': ['Arizona 1 2014-12-23       3242.0',
                'Iowa 1 2010-02-23       3453.7',
                'Oregon 0 2014-06-20       2123.0',
                'Maryland 0 2014-03-14       1123.6',
                'Florida 1 2013-01-15       2134.0',
                'Georgia 0 2012-07-14       2345.6']}
```

```python
df = pd.DataFrame(data, columns=["raw"])
```

```python
df
```

```python
df["Female"] = df["raw"].str.extract("(\d)")
```

```python
df["Date"] = df["raw"].str.extract("(\d\d\d\d-\d\d-\d\d)")
```

```python
df["State"] = df["raw"].str.extract("([A-Z]\w{1,})")
```

```python
df["State"]
```

```python
df["Score"] = df.raw.str.extract("(\d\d\d\d\.\d)")
```

```python
df["Score"]
```

```python
titanic["Title"] = titanic.Name.str.extract("([A-Z]\w+)\.")
```
